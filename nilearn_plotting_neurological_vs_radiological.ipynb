# nilearn neurological view vs radiological view example

This simple notebook was created to explore plotting in neurological view versus radiological view in nilearn.

**NOTE:** Radiological view has not yet been implemented in nilearn as of version 0.5.0. An issue has been opened to implement this enhancement properly in a future version of nilearn: https://github.com/nilearn/nilearn/issues/1875

A workaround has described on [neurostars](https://neurostars.org/t/how-to-control-displayed-orientation-in-nilearn-plots/1408) using the `nilearn.image.swap_img_hemispheres()` function. However, this should be used with extreme caution and of course for research purposes only.

```python
from nilearn import plotting, image, datasets
```

```python
mni_img = datasets.load_mni152_template()
swap_mni_img = image.swap_img_hemispheres(mni_img) # CAUTION: left-right swapped(!)

plotting.plot_anat(mni_img, title="Neurological View", cut_coords=[5, 0, 0], draw_cross = True)
plotting.plot_anat(swap_mni_img, title="L/R Swapped: CAUTION!", cut_coords=[5, 0, 0], draw_cross = True)
plotting.plot_anat(swap_mni_img, title="L/R Swapped (No Labels): CAUTION!", cut_coords=[5, 0, 0], draw_cross = True, annotate = False)
```
